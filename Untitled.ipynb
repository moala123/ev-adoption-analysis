## Data Cleaning & Preprocessing
In this section, missing values were reviewed and handled where appropriate
to ensure data quality before analysis.


```python
for col in ['City', 'County', 'Postal Code', 'Electric Utility']:
    df[col] = df[col].fillna(df[col].mode()[0])
```

Columns with a high percentage of missing values, such as Base MSRP and Electric Range,
were intentionally left unchanged, as these values are not mandatory during data collection
and imputing them could introduce bias.


### Cleaning Electric Range
Zero values in the 'Electric Range' column are treated as missing values (`NaN`) 
because they do not represent real-world driving distances.


```python
df["Electric Range"] = df["Electric Range"].replace(0, np.nan)
```

### Cleaning Base MSRP
Base MSRP values equal to zero are treated as missing, as they would skew any price-related analysis.


```python
df["Base MSRP"] = df["Base MSRP"].replace(0, np.nan)
```

### CAFV Eligibility Simplification
CAFV eligibility values are simplified into three categories:  
- **Eligible**  
- **Not Eligible**  
- **Unknown**  

This makes analysis easier and more interpretable.


```python
def simplify_cafv(status):
    if "Eligible" in str(status):
        return "Eligible"
    elif "Not eligible" in str(status):
        return "Not Eligible"
    else:
        return "Unknown"

df["CAFV Status"] = df["Clean Alternative Fuel Vehicle (CAFV) Eligibility"].apply(simplify_cafv)
```

### Electric Vehicle Types
Battery Electric Vehicles (BEVs) are more prevalent than Plug-in Hybrid Electric Vehicles (PHEVs).  
This reflects the market's preference toward fully electric vehicles.


```python
df["Electric Vehicle Type"].value_counts()
```

### Top 10 Manufacturers
Tesla is the dominant manufacturer in the electric vehicle market, followed by Nissan, Chevrolet, Ford, and Toyota.  
Understanding the leading manufacturers helps identify market trends.


```python
top_makes = df["Make"].value_counts().head(10)
top_makes
```

### Vehicle Registrations Over the Years
EV adoption increased significantly after 2018, with rapid growth post-2020.  
This indicates a strong market expansion and adoption of electric vehicles in recent years.


```python
year_counts = df["Model Year"].value_counts().sort_index()
year_counts
```

### Average Electric Range by Vehicle Type
Battery Electric Vehicles (BEVs) offer substantially higher ranges than Plug-in Hybrid Electric Vehicles (PHEVs),  
making them more suitable for longer trips.


```python
range_by_type = df.groupby("Electric Vehicle Type")["Electric Range"].mean()
range_by_type
```

### Average Base MSRP by Vehicle Type
BEVs tend to have higher base prices compared to PHEVs.  
This correlates with their higher electric range and advanced technology features.


```python
price_by_type = df.groupby("Electric Vehicle Type")["Base MSRP"].mean()
price_by_type
```

### Distribution of Electric Vehicle Types
This bar chart shows that Battery Electric Vehicles (BEVs) are more common than Plug-in Hybrid Electric Vehicles (PHEVs), 
indicating stronger adoption of fully electric vehicles.


```python
plt.figure(figsize=(6,4))
df["Electric Vehicle Type"].value_counts().plot(kind="bar")
plt.title("Distribution of Electric Vehicle Types")
plt.xlabel("Vehicle Type")
plt.ylabel("Count")
plt.show()
```

### Top EV Manufacturers
Tesla clearly dominates the market, followed by a group of traditional automakers.
This highlights Teslaâ€™s strong position in the EV ecosystem.


```python
plt.figure(figsize=(8,5))
top_makes.plot(kind="bar")
plt.title("Top 10 EV Manufacturers")
plt.xlabel("Manufacturer")
plt.ylabel("Number of Vehicles")
plt.show()
```

### EV Adoption Over Time
The line chart shows a sharp increase in EV registrations after 2020, 
indicating rapid market growth and increased adoption of electric vehicles.


```python
plt.figure(figsize=(8,5))
year_counts.plot()
plt.title("EV Registrations Over Time")
plt.xlabel("Model Year")
plt.ylabel("Number of Vehicles")
plt.show()
```

### Electric Range Distribution
Most vehicles have an electric range below 300 miles.
Higher ranges are less common and are mostly associated with newer BEV models.


```python
plt.figure(figsize=(8,5))
df["Electric Range"].dropna().hist(bins=40)
plt.title("Distribution of Electric Range")
plt.xlabel("Electric Range (miles)")
plt.ylabel("Frequency")
plt.show()
```

### Electric Range by Vehicle Type
BEVs show significantly higher and more variable electric ranges compared to PHEVs, 
which typically have short electric-only ranges.


```python
plt.figure(figsize=(6,4))
sns.boxplot(
    data=df,
    x="Electric Vehicle Type",
    y="Electric Range"
)
plt.title("Electric Range by Vehicle Type")
plt.show()
```

### Base MSRP by Vehicle Type
BEVs generally have higher base prices than PHEVs.
Price variability is also higher among BEVs, reflecting a wider range of models and technologies.


```python
plt.figure(figsize=(6,4))
sns.boxplot(
    data=df,
    x="Electric Vehicle Type",
    y="Base MSRP"
)
plt.title("Base MSRP by Vehicle Type")
plt.show()
```

### Top Counties for EV Adoption
King County leads EV adoption by a large margin, followed by Snohomish and Pierce counties.
Urban and high-income areas show higher EV adoption rates.


```python
top_counties = df["County"].value_counts().head(10)

plt.figure(figsize=(8,5))
top_counties.plot(kind="bar")
plt.title("Top 10 Counties by EV Registrations")
plt.xlabel("County")
plt.ylabel("Number of Vehicles")
plt.show()
```
